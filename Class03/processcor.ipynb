```python
import pandas as pd
import numpy as np
```

```python
headers = [
    'login_session_id',
    'login_timestamp',
    'user',
    'logout_timestamp',
    'mac_address',
    'ipv4',
    'ipv6',
    'agent_ip',
    'agent_type',
    'via_ip',
    'ipv4_byte_in',
    'ipv4_byte_out',
    'ipv4_pkt_in',
    'ipv4_pkt_out',
    'ipv6_byte_in',
    'ipv6_byte_out',
    'ipv6_pkt_in',
    'ipv6_pkt_out',
    'last_seen_timestamp'
    ]

login_data = pd.read_csv('login-20170102-anon.csv.txt', sep=' ', names=headers)
login_data
```

```python
def to_datetime(data):
    tz = pd.Timedelta(hours=7)
    data['login_timestamp'] = pd.to_datetime(data['login_timestamp'], unit='ms') + tz
    data['logout_timestamp'] = pd.to_datetime(data['logout_timestamp'], unit='ms', errors='coerce')  + tz
    data['last_seen_timestamp'] = pd.to_datetime(data['last_seen_timestamp'], unit='ms')  + tz
    return data

login_data = to_datetime(login_data)
```

```python
with_logout_ts = login_data[~pd.isnull(login_data['logout_timestamp'])]

sorted_login = with_logout_ts.sort_values(by=['login_timestamp'])

RANGE = 2
LOWER_BOUND = pd.Timestamp('2017-01-01')
UPPER_BOUND = LOWER_BOUND + pd.Timedelta(days=RANGE)

outlier = list(sorted_login.index[sorted_login['login_timestamp']<LOWER_BOUND])
outlier += list(sorted_login.index[sorted_login['login_timestamp']>UPPER_BOUND])
sorted_login = sorted_login.drop(outlier).reset_index(drop=True)
sorted_login
```

```python
start_ts = sorted_login['login_timestamp'][0].floor('15min')
end_ts = sorted_login.tail(1)['login_timestamp'].reset_index(drop=True)
end_ts = end_ts[0].ceil('15min')
start_ts, end_ts
```

```python
import collections

time_axis = {}
step = pd.Timedelta(minutes=15)
iterator = start_ts

while iterator < end_ts:
    time_axis[iterator] = 0
    iterator += step

time_axis = collections.OrderedDict(sorted(time_axis.items()))
user_number = time_axis.copy()
len(user_number)
```

```python
ticks = list(user_number.keys())

def get_inner(data, i):
    left_outer = data['logout_timestamp'] < ticks[i]
    right_outer = data['login_timestamp'] > ticks[i+1]
    return data[~(left_outer | right_outer)]

for i in range(len(ticks)-1):
    user_number[ticks[i+1]] = get_inner(sorted_login, i)['user'].nunique()

user_number
```

```python
import json

def get_dict_str(d):
    dict_str = collections.OrderedDict()
    for k, v in d.items():
        dict_str[str(k)] = v
    return dict_str

user_number_str = get_dict_str(user_number)
```

```python
with open('user-number.json', 'w') as f:
    json.dump(user_number_str, f, indent=4, sort_keys=True)
```

```python
ipv4_number = time_axis.copy()

for i in range(len(ticks)-1):
    inner = get_inner(sorted_login, i)
    ipv4 = inner['ipv4']!='-'
    ipv6 = inner['ipv6']!='-'
    ipv4_number[ticks[i+1]] = inner[ipv4 & ~ipv6]['ipv4'].nunique()
    
ipv4_number
```

```python
ipv6_number = time_axis.copy()

for i in range(len(ticks)-1):
    inner = get_inner(sorted_login, i)
    ipv4 = inner['ipv4']!='-'
    ipv6 = inner['ipv6']!='-'
    ipv6_number[ticks[i+1]] = inner[~ipv4 & ipv6]['ipv6'].nunique()
    
ipv6_number
```

```python
dualstack_number = time_axis.copy()

for i in range(len(ticks)-1):
    inner = get_inner(sorted_login, i)
    ipv4 = inner['ipv4']!='-'
    ipv6 = inner['ipv6']!='-'
    dualstack_number[ticks[i+1]] = inner[ipv4 & ipv6]['ipv4'].nunique()
    
dualstack_number
```

```python
with open ('ipv4-number.json', 'w') as f:
    json.dump(get_dict_str(ipv4_number), f, sort_keys=True, indent=4)

with open ('ipv6-number.json', 'w') as f:
    json.dump(get_dict_str(ipv6_number), f, sort_keys=True, indent=4)

with open ('dualstack-number.json', 'w') as f:
    json.dump(get_dict_str(dualstack_number), f, sort_keys=True, indent=4)
```

```python
def get_between(data, i):
    login = (ticks[i] < data['login_timestamp']) & (data['login_timestamp'] <= ticks[i+1])
    logout = (ticks[i] < data['logout_timestamp']) & (data['logout_timestamp'] <= ticks[i+1])
    return data[login].shape[0], data[logout].shape[0]
```

```python
login_count = time_axis.copy()
logout_count = time_axis.copy()

for i in range(len(ticks)-1):
    login_count[ticks[i+1]], logout_count[ticks[i+1]] = get_between(sorted_login, i)
```

```python
with open('login-count.json', 'w') as f:
    json.dump(get_dict_str(login_count), f, sort_keys=True, indent=4)
    
with open('logout-count.json', 'w') as f:
    json.dump(get_dict_str(logout_count), f, sort_keys=True, indent=4)
```
