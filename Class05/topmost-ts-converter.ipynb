```python
import pandas as pd
import json
```

```python
f = open('csv/topmost-5-user.csv', 'r')
headers = ['request_timestamp', 'user_id_client', 'src_ip ', 'dst_ip', 'dst_port', 'host']
data = pd.read_csv(f, sep=',', names=headers)
data
```

```python
data['request_timestamp'] = pd.to_datetime(data['request_timestamp']//1000000, unit='s') + pd.Timedelta('7 hours')
```

```python
data['request_timestamp'] = pd.to_datetime(data['request_timestamp'],format= '%H:%M:%SS' ).dt.time
data['dst_port'] = data['dst_port'].astype(str)
```

```python
import csv
with open('csv/topmost-ts-converted.csv', 'w') as f:
    data.to_csv(f, sep=',', header=False, index=False)
```
