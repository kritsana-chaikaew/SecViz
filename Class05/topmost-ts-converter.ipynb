```python
import pandas as pd
import json
```

```python
f = open('topmost-user.csv', 'r')
headers = ['request_timestamp', 'user_id_client', 'src_ip ', 'dst_ip', 'dst_port', 'host']
data = pd.read_csv(f, sep=',', names=headers)
data
```

```python
data['request_timestamp'] = pd.to_datetime(data['request_timestamp'], unit='us') + pd.Timedelta('7 hours')
```

```python
with open('topmost-converted-ts.csv', 'w') as f:
    data.to_csv(f, sep=',', header=None, index=None)
```
